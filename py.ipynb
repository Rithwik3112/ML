```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

X = np.array([1, 2, 4, 6, 7])
Y = np.array([2, 3, 5, 2, 3])
```

```python

m = 0.0 
c = 0.0  
alpha = 0.01 
epochs = 1000  
n = len(X)  
```

```python
for _ in range(epochs):
    Y_pred = m * X + c
    error = Y - Y_pred

    # Compute gradients
    dm = (-2 / n) * np.sum(X * error)
    dc = (-2 / n) * np.sum(error)

    # Update parameters
    m = m - alpha * dm
    c = c - alpha * dc
```

```python
print(f"Slope (m): {2:.2f}")
print(f"Intercept (c): {c:.2f}")

```

```python
plt.scatter(X, Y, marker='o')
plt.plot(X, Y_pred, color='red')
plt.xlabel('feature')
plt.ylabel('lable')

```
